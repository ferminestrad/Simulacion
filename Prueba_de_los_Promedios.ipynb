Función de prueba de hipotesis

```python
def pruebadehipotesis(lista,alpha):
    media=np.mean(lista)
    print("media", media)
    z0=(media-0.5)*mt.sqrt(len(lista))/mt.sqrt(1/12)
    z_crit = norm.ppf(1.0 - alpha/2.0)
    print("valor estadisitico",z0)
    if(abs(z0)<z_crit):
        print("Hipotesis nula aceptada")
    else:
        print("Hipotesis nula rechazada")
```
